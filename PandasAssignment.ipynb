````python
# 1 - shape of dataset

import pandas as pd

df = pd.read_csv("data.csv")

shape = df.shape

print("Shape = {}".format(shape))
````

````python
# 2- first 15 rows of dataset
import pandas as pd
df = pd.read_csv("data.csv")
first_15_rows = df[:15]
print(first_15_rows)
````

````python
# 3- information for missing values in certain collums

import pandas as pd
data = pd.read_csv("data.csv")
bool_series = pd.notnull(data["FuelEconomy"])
bool_series2 = pd.notnull(data["Comments"])
print (bool_series)
print (bool_series2)
````

````python

# 4- deleting duplicates

import pandas as pd
data = pd.read_csv("data.csv")

def drop_duplicates(df, subset=None, keep='last', inplace=False, ignore_index=False):
    result = df.drop_duplicates(subset=subset, keep=keep, inplace=inplace, ignore_index=ignore_index)
    if not inplace:
        return result


df_no_duplicates = drop_duplicates(data)
print(df_no_duplicates)
````

````python
# 5 - Number of missing values in the rows

import pandas as pd

data = pd.read_csv("data.csv")
missing_values_count = data.isnull().sum()
print(missing_values_count)
````

````python
# 6 - 2 copies of dataframe with tweaks

import pandas as pd
data = pd.read_csv("data.csv")

data_no_missing_rows = data.dropna()
data_no_missing_columns = data.dropna(axis=1)

print("DataFrame without missing rows:")
print(data_no_missing_rows)
print("\nDataFrame without missing columns:")
print(data_no_missing_columns)
````

````python
# 7 - summary statistics and correlation
import pandas as pd
data = pd.read_csv("data.csv")

summary_statistics = data.describe()

print("Summary Statistics:")
print(summary_statistics)

numerical_collums = ['Distance', 'MaxSpeed', 'AvgSpeed', 'AvgMovingSpeed', 'TotalTime', 'MovingTime']
correlation_matrix = data[numerical_collums].corr()


print("Correlation Matrix:")
print(correlation_matrix)
````

````python
# 8 - rows 11 and 14

import pandas as pd
data = pd.read_csv("data.csv")

print("Row 11:")
print(data.iloc[10])

print("Row 14:")
print(data.iloc[13])

````

````python
# 9 - subsets

import pandas as pd
data = pd.read_csv("data.csv")

subset_nov = data[data['Date'] == '11/23/2011']
subset_jan = data[data['Date'] == '1/6/2012']

print(subset_nov)
print('\n')
print(subset_jan)

````

````python
# 10 - scatter plot

import pandas as pd
data = pd.read_csv("data.csv")

import matplotlib.pyplot as plt
plt.rcParams.update({'font.size': 20, 'figure.figsize': (10, 8)})

plt.scatter(data['Distance'], data['TotalTime'])

plt.title('Distance vs Total Time')
plt.xlabel('Distance')
plt.ylabel('Total Time')

plt.show()
````

````python
# 11 - boxplot

import matplotlib.pyplot as plt
import pandas as pd
data = pd.read_csv("data.csv")

plt.rcParams.update({'font.size': 20, 'figure.figsize': (10, 8)})
data[['AvgSpeed', 'AvgMovingSpeed']].boxplot()


plt.title('Average Speed vs Average Moving Speed')
plt.ylabel('Speed')


plt.show()
````
